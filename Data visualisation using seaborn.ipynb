```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
df = pd.read_csv('C:/Users/afsaln2/Desktop/kaggle/icfos/Assignments/asg 3/citizen.csv')
print(df.head())
```

```python
sns.countplot(x='sex',data=df)
#Shows that there are more number of males in the dataset
plt.figure(figsize=(20,5))


sns.countplot(x='native.country', data=df)
#Shows that large majority of the people have USA as native country
```

```python
plt.figure(figsize=(20,5))
sns.regplot(x='age', y='hours.per.week', data=df)
#There seems to be no meaningful correlation between hours of work and age
```

```python
sns.boxplot(x="sex", y="hours.per.week", data=df,palette='rainbow')
#Here, we find that males work for more hours per week than females
```

```python
plt.figure(figsize=(20,5))

sns.boxplot(x="sex", y="hours.per.week", data=df,palette='rainbow',hue='education', orient="v")
```

```python
sns.countplot(x='education',hue = 'sex', data=df)
#Shows that in all categories males are more educated than females
```

```python
sns.stripplot(x="sex", y="age", data=df)
```
